```python
import numpy as np
import keras
from PIL import Image
```

```python
import glob
```

```python
train_files = glob.glob("C:/Users/Rohan/OneDrive/Desktop/is_the_road_safe/assignment_2/chest_xray/train/*/**")
```

```python
from random import shuffle
```

```python
shuffle(train_files)
train_files
```

```python
def give_index(file_name): 
    if((file_name.find("NORMAL"))!=-1):
        return 1 
    
    elif((file_name.find("virus"))!=-1):
        return 2
    
    
    
    elif((file_name.find("bacteria"))!=-1):
        return 3        
    
    else:
        return 0
```

```python
X_train = np.zeros((len(train_files),128,128))
Y_train = np.zeros((len(train_files)))

for i,fl in enumerate(train_files):
    img = Image.open(fl)
    
    img = img.resize((128,128))
    img = img.convert("L")
    X_train[i,:,:] = np.array(img)
    Y_train[i] = give_index(fl)
```

```python
Y_train
```

```python
X_train=X_train/255.0
```

```python
import keras
```

```python
model = keras.Sequential([
    
    keras.layers.Conv2D(
    filters=120,
    kernel_size=3,
        # filtering out negative values
    activation='relu',
    input_shape=(128,128,1)
    ),

    keras.layers.Conv2D(
    filters=24,
    kernel_size=3,
    activation='relu',
    ),

    # 1d matrix
    keras.layers.Flatten(),
    # 4
    # reduction layers
    keras.layers.Dense(
        units=96,
        activation='relu',
    ),

    # 96 units -> 4 units
    keras.layers.Dense(4,activation='softmax') 
    
])
```

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-3),
              loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
X_train=X_train.reshape((len(train_files),128,128,1))
```

```python
model.fit(X_train,Y_train,epochs=2,validation_split=0.1)
```
